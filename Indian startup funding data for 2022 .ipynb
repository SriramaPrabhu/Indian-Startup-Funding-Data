```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url='https://startuptalky.com/indian-startups-funding-investors-data-2022/#Indian_Startups_Funding_Data_-_January_2022'
allcontent=requests.get(url)
soup=BeautifulSoup(allcontent.text,'html.parser')
# print(soup.prettify())

data=[]

monthdata=soup.find_all('div','content','table')
for month in monthdata:
    info=month.find_all('tbody')
    for inf in info:
        indivdata=inf.find_all('tr')
        for indiv in indivdata:
            columns=indiv.find_all('td')
            name=columns[0].text
            Sector=columns[3].text
            Amount=columns[7].text
            FundingType=columns[8].text
            HQ=columns[2].text
            Investor=columns[6].text
            data.append([name,Sector,Amount,FundingType,HQ,Investor])
df=pd.DataFrame(data,columns=['Name','Sector','Amount Raised (USD)','FundingType','Headquarter','Investor'])
df
```
